Write a program that asks the user how many Fibonacci numbers to generate and
then generates them. Make sure to ask the user to enter the number of numbers in the
sequence to generate.

```python
num=int(input("HOW MANY FIBONACCI NUMBER WOULD YOU LIKE TO GENERATE ? "))
```

```python
f0 = 0 
f1 = 1
print(f1,end=',')
for i in range(num):
    f = f0 + f1
    f0 = f1
    f1 = f
    print(f,end=',')
```
